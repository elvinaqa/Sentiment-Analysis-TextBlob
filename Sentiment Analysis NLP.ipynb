```
from textblob import TextBlob
import nltk
from newspaper import Article
```

```
url = 'https://medium.com/@elvinaqa/deep-learning-and-computer-vision-basics-i-86acab09e3b7'
```

```
article = Article(url)
```

```
article.download()
article.parse()
```

```
article.nlp()
```

```
text = article.summary
print(text)
```

```
to = TextBlob(text)
sentiment = to.sentiment.polarity
print(sentiment)
```
